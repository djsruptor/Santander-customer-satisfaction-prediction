```python
import requests
import json
import os
from src.data import get_data, remove_correlated_features, json_customer
```

```python
customer = 5

if not os.path.exists(f'examples/{customer}.json'):
    print(f'File not yet created, use the GET option to create {customer}.json')
else:
    with open(f'examples/{customer}.json', 'r') as f:
        data = json.load(f)

    url = f'http://localhost:9696/predict?number={customer}'

    response = requests.post(url, json=data)
    dissatisfaction = response.json()

    print('Response: ', dissatisfaction)

    if dissatisfaction['dissatisfied'] > 0.136:
        print('Customer is leaving')
    else:
        print('Customer is happy')
```
